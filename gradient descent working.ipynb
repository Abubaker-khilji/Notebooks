```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples = 4, n_features=1 , n_informative=1 , n_targets=1 , noise=80 , random_state=13) 
```

```python
plt.scatter(X, y)
```

```python
#ok first we apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg =  LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color="red")
```

```python
#lets apply Gradient Descent assuimng slope is constant =78.35
#and assume the starting value for intercept b = 0
y_pred = ((78.35*X) + 100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X) , color = "red", label =  "OLS")
plt.plot(X, y_pred, color="black" , label = "b = 0")
plt.legend()
plt.show()
```

```python
m = 78.35
b_old = 100
```

```python
loss_slope = -2 * np.sum(y - m*X.ravel()-b_old)
```

```python
loss_slope
```

```python
#we begin by making our lr means learning rate as = 0.1
lr =  0.1
```

```python
step_size = loss_slope*lr
step_size
```

```python
#CALCULATING THE NEW INTERCEPT 
b_new = b_old -  step_size
b_new
```

```python

y_pred1 = ((78.35 * X) + b_new).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='yellow',label='b = {}'.format(b_new))
plt.plot(X,y_pred,color='black',label='b = 0')
plt.legend()
plt.show()
```

```python
#interation 2
loss_slope = -2 * np.sum(y - m*X.ravel()-b_new)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b_new_new = b_new  - step_size
b_new_new
```

```python

y_pred2 = ((78.35 * X) + b_new_new).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='pink',label='b = {}'.format(b_new))
plt.plot(X,y_pred1,color='green',label='b = {}'.format(b_new_new))
plt.plot(X,y_pred,color='black',label='b = 0')
plt.legend()
plt.show()
     
```

```python
#iteration 3 
loss_slope = -2 * np.sum(y - m*X.ravel()-b_new_new)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b_new_new_new = b_new_new - step_size
b_new_new_new
```

```python
y_pred3 = ((78.35 * X) + b_new_new_new).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='purple',label='b = {}'.format(b_new_new_new))
plt.plot(X,y_pred2,color='pink',label='b = {}'.format(b_new))
plt.plot(X,y_pred1,color='green',label='b = {}'.format(b_new_new))
plt.plot(X,y_pred,color='black',label='b = 0')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
    loss_slope = -2 *np.sum(y - m*X.ravel() - b)
    b = b -(lr * loss_slope)
    y_pred = m * X + b
    plt.plot(X,y_pred)

plt.scatter(X,y)
```
