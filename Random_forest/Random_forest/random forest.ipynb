```python
#Numbers of trees in random forest 
n_estimators = [20,60,100,120]
# Numbers of features aka columns to be consider at every split
max_features = [0.2,0.6,1.0]
# Maximu, numbers of levels in tree
max_depth = [2,8,None]
#Numbers of samples
max_samples = [0.5,0.75,1.0]
```
